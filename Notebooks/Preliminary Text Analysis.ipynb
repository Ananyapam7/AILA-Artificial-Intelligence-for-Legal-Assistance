```python
import os # accessing directory structure
import pandas as pd
```

```python
f = open("AILA_2019_dataset/Object_casedocs/C1.txt", "r")
text_doc = f.read() 
print(text_doc) #The legal text document
```

```python
import re
pattern = r"[^\w]" #replaced the non-words with spaces
text = re.sub(pattern, " ", text_doc)
print(text)
```

```python
import nltk
from nltk.corpus import stopwords 
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(text)
for sentence in sentences:
    print(sentence)
    print()
```

```python
for sentence in sentences:
    words = nltk.word_tokenize(sentence) #tokenize
    print(words)
    print()
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer() # creating a Porterstemmer object for stemming
```

```python
stemmed_text = stemmer.stem(text)

print(stemmed_text)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english")) # Removed the stop-words

words = nltk.word_tokenize(stemmed_text)
text_without_stop_words = [word for word in words if not word in stop_words]
print(text_without_stop_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer #TF-IDF
from sklearn.feature_extraction.text import CountVectorizer #Bag-of-words

count_vectorizer = CountVectorizer()

bag_of_words = count_vectorizer.fit_transform(text_without_stop_words)

feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

```python
tfidf_vectorizer = TfidfVectorizer()
values = tfidf_vectorizer.fit_transform(text_without_stop_words)

feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(values.toarray(), columns = feature_names)
```
